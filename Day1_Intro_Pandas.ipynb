###**Grouped Summaries .groupby(), .apply() and .agg() method (Advance Topic)**




```
grouped_df1 = avocad.groupby(['year','type'])[['AveragePrice','Total Volume','Total Bags']].apply(np.mean)
print('Apply method:\n')
grouped_df1


# for key, item in grouped_df:
#   print(grouped_df.get_group(key)
```

```
def pct30(column):     
    #return the 0.3 quartile
    return column.quantile(0.3)

def pct75(column):     
    #return the 0.5 quartile
    return column.quantile(0.75)
    
grouped_df2 = avocad.groupby(['year','type'])[['AveragePrice','Total Volume','Total Bags']].agg([min,max,np.mean,np.median,pct30,pct75])
print('Aggregate method:\n')
grouped_df2
```

###**Handling Missing Value with Pandas**
Source of missing values:


*   User forgot to fill in a field
*   Data lost during transfer/copy
*   Systematic errors (during logging)







```
prop_df = pd.read_csv('/content/drive/My Drive/data/property_data.csv')
prop_df
```

```
# check for missing value:
print(prop_df.isna().any())

print('\n',prop_df.isna().sum())
```

```
# Making a list of missing value types
missing_values = ["n/a", "na", "--"]
prop_df = pd.read_csv('/content/drive/My Drive/data/property_data.csv', na_values = missing_values)
prop_df

#prop_df.replace({'n/a':np.nan,'--': np.nan, 'na':np.nan}, inplace = True)
```

```
prop_df.loc[3,'OWN_OCCUPIED'] = 'Y'
prop_df.loc[6,'NUM_BATH'] = np.nan
prop_df
```

###**Missing Value Imputation**

*   Drop missing value
*   Replace missing value



```
# drop missing value
# drop if any cell contain NaN
df2 = prop_df.copy()
df2.dropna(axis=0, how='any', inplace=True)
df2
```

```
# drop only if all cells in the row contain NaN
df3 = prop_df.copy()
df3.dropna(axis=0, how= 'all', inplace=True)
df3
```

```
# drop row if the row contain at least 3 missing value
# thresh: number of non-NA values
df4 = prop_df.copy()
df4.dropna(axis=0, thresh=4, inplace=True)
df4
```

```
# replacing missing values:
# fillna(x)

df5 = prop_df.copy()
df5.fillna(-1, inplace=True)
df5
```

```
# limit: set number of missing value to be forward or backward filled
# method: 'ffill' and 'bfill'
df6 = prop_df.copy()
df6.fillna(axis=0, method='ffill',limit=1, inplace=True)
df6
```
